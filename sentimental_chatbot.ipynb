```python
openai.api_key = "******************NjEWcvEdpQodyvXZxFO"  # Replace with your actual API key


sia = SentimentIntensityAnalyzer()


def analyze_sentiment(message):
    scores = sia.polarity_scores(message)
    if scores['compound'] >= 0.05:
        return "positive"
    elif scores['compound'] <= -0.05:
        return "negative"
    else:
        return "neutral"

# Initialize the Hugging Face pipeline for text generation
generator = pipeline('text-generation', model='gpt2') # You can change the model here

# Function to generate responses using Hugging Face
def generate_response(user_input, length=100):
    prompt = f"User: {user_input}\nChatbot:"
    generated_text = generator(prompt, max_length=length, num_return_sequences=1)
    response = generated_text[0]['generated_text'].replace(prompt, "").strip()
    return response

# Chatbot interaction loop
def chatbot():
    print("Chatbot: Hi! I'm here to help. Type 'exit' to end the chat.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            print("Chatbot: Goodbye! Have a great day! ðŸ˜Š")
            break

        sentiment = analyze_sentiment(user_input)
        response = generate_response(user_input)

        if sentiment == "positive":
            print(f"Chatbot: I'm glad to hear that! ðŸ˜Š {response}")
        elif sentiment == "negative":
            print(f"Chatbot: I'm sorry you're feeling this way. {response} ðŸ™")
        else:
            print(f"Chatbot: Thanks for sharing. {response} ðŸ˜Š")

# Run the chatbot
if __name__ == "__main__":
    chatbot()
```
